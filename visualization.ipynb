```python
import matplotlib.pyplot as plt
import re
```

```python
def load_execution_times():
    times = {}
    try:
        with open("output\\report.txt", "r") as file:
            content = file.readlines()
            for line in content:
                match = re.match(r"\s*(\d+)\s*\|\s*(\d+)", line)
                if match:
                    size = int(match.group(1))
                    time = int(match.group(2))
                    times[size] = time
    except Exception as e:
        print(f"Error loading execution times: {e}")

    return times


def plot_execution_times(times):
    sizes = sorted(times.keys())
    exec_times = [times[size] for size in sizes]

    plt.figure(figsize=(8, 5))
    plt.plot(sizes, exec_times, marker='o', linestyle='-',
             color='b', label="Execution Time")
    plt.xlabel("Matrix Size (NxN)")
    plt.ylabel("Time (ms)")
    plt.title("Execution Time X Matrix Size")
    plt.grid(True)
    plt.legend()
    plt.show()


execution_times = load_execution_times()
if execution_times:
    plot_execution_times(execution_times)
```
